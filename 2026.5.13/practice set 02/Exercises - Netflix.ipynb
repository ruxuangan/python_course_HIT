7- Create __a dictionary__ where the titles are the keys and the cast names are the values.

<span style="color:crimson">Note:</span> Only add titles where the cast is not null. Drop rows where the cast is Null.

<span style="color:crimson">Note:</span> Name this dictionary as __titles_and_cast__.

```python
# your code
mask = ~ df["cast"].isnull()
df_cast = df[mask]
titles_and_cast = {}
for i, row in df_cast.iterrows():
    titles_and_cast[row["title"]]= row["cast"]

```

8- Using the __titles_and_cast__ dictionary, write a piece of code to find out how many times "Ryan Reynolds" played in a movie or a TV show. Print the titles.

```python
# your code
for k,v in titles_and_cast.items():
    names = v.split(",")
    for n in names:
        if n.strip().lower()== "Ryan Reynolds".lower():
            print(k)
```

9- Using the __titles_and_cast__ dictionary, write a piece of code to find out how many times "Ryan Reynolds" and "Blake Lively" played in a movie or a TV show together. Print the titles.

```python
# your code
for k,v in titles_and_cast.items():
    names = v.split(",")
    for n in names:
        if n.strip().lower()== "Ryan Reynolds".lower():
                for n in names:
                    if n.strip().lower()== "Blake Lively".lower():
                         print(k)
```
