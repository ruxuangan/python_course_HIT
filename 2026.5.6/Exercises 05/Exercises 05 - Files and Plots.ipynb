__Text Files__

You are given a __.txt__ file called "poem.txt".

Write a script that reads in the file and prints the poem.

```python
# your code
my_file = open(file = "poem.txt", mode = "r")
print(my_file.read())
my_file.close()
```

The following sentence is missing from the end of the poem:

__If he believes in equality.__

Add it to the end of the file and save it. Check your file to make sure you did it correctly.

```python
# your code
my_file = open(file = "poem.txt", mode = "a")
my_file.write("\nIf he believes in equality.")
my_file.close()
my_file = open(file = "poem.txt", mode = "r")
print(my_file.read())
```

__Log Files__

You are given a .txt file called __mbox-short.txt__.

Write a code that reads the file and looks for lines containing 'http'. Your task is to extract the URLs from these lines and print them out.


```python
myfile = open("mbox-short.txt", "r")
out = myfile.readlines()
myfile.close()

# your code
for line in out:
    if line[:4] == "http":
        print(line)
```
