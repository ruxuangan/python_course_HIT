### Homework 02 - Personal Library

Create a Python class that allows users to manage their book collection.

Keep the following information for each book:

    Title (string)
    Author (string)
    Publication Year (integer)
    Genre (string)
    Read Status (boolean: True if read, False if unread)


Add the following methods to your class and test them:

__Add a Book:__- Prompt the user to enter the book's details and add it to the library.

__Remove a Book:__ Prompt the user to enter the title of the book to remove it from the library.

__Search for a Book:__ Allow the user to search for a book by title or author. Display all matching books.

__Display All Books:__ Show all books in the library in a formatted way.

__Display Statistics:__ Print: Total number of books in the library and Percentage of books that have been read.

__Bonus:__

__Save:__ Save the library data to a file (e.g., library.txt) 

---

```python
import pandas as pd

# create the file if there is none
my_file = open(file = "library.txt", mode = "a")
my_file.close()

class Library:
    def __init__(self, title, author, publication_year, genre, read_status):
        self.title = title
        self.author = author
        self.publication_year = publication_year
        self.genre = genre
        self.read_status = read_status
    
    def add_a_book(self):
        my_file = open(file = "library.txt", mode = "a")
        my_file.write(f"{self.title}, {self.author}, {self.publication_year}, {self.genre}, {self.read_status}\n")
        my_file.close()
        print("successfully add a book!")

    def remove_a_book(self):
        my_file = open(file = "library.txt", mode = "r")
        lines = my_file.readlines()
        my_file.close()
        removed_book = []
        remained_book = []
        for line in lines:
            one_book = [item.strip() for item in line.split(",")]
            if self.title == one_book[0].strip():
                removed_book.append(one_book)
            else:
                remained_book.append(line)
        if len(removed_book) == 0:
            print(f"There is no book titled {self.title}")
        else:
            df = pd.DataFrame(removed_book, columns = ['title', 'author', 'publication_year', 'genre', 'read_status'])
            print(df)
            check = input("Do you want to remove the printed book? Enter yes or no")
            if check == "yes":
                my_file = open(file = "library.txt", mode = "w")
                my_file.writelines(remained_book)
                my_file.close()
                print("Successfully remove the book!")
            elif check == "no":
                print("Stop removing the book")
            else:
                print("Please enter yes or no")   

    def search_for_a_book_title(self):
        my_file = open(file = "library.txt", mode = "r")
        lines = my_file.readlines()
        my_file.close()
        search_book = []
        for line in lines:
            one_book = [item.strip() for item in line.split(",")]
            if self.title == one_book[0]:
                search_book.append(one_book)
        if len(search_book) == 0:
            print(f"There is no book titled {self.title}")
        else:
            print("All matching books are as follow")
            df = pd.DataFrame(search_book, columns = ['title', 'author', 'publication_year', 'genre', 'read_status'])
            print(df)

    def search_for_a_book_author(self):
        my_file = open(file = "library.txt", mode = "r")
        lines = my_file.readlines()
        my_file.close()
        search_book = []
        for line in lines:
            one_book = [item.strip() for item in line.split(",")]
            if self.author == one_book[1].strip():
                search_book.append(one_book)
        if len(search_book) == 0:
            print(f"There is no book titled {self.author}")
        else:
            print("All matching books are as follow")
            df = pd.DataFrame(search_book, columns = ['title', 'author', 'publication_year', 'genre', 'read_status'])
            print(df)

def display_all_books():
    my_file = open(file = "library.txt", mode = "r")
    lines = my_file.readlines()
    my_file.close()
    display_book = []
    for line in lines:
        one_book = [item.strip() for item in line.split(",")]
        display_book.append(one_book)
    if len(display_book) == 0:
        print("No book in your library!")
    else:
        print("All books are as follow")
        df = pd.DataFrame(display_book, columns = ['title', 'author', 'publication_year', 'genre', 'read_status'])
        print(df)

def display_statistic():
    my_file = open(file = "library.txt", mode = "r")
    lines = my_file.readlines()
    my_file.close()
    display_statistic = []
    read_book = 0
    for line in lines:
        one_book = [item.strip() for item in line.split(",")]
        display_statistic.append(one_book)
        if one_book[4] == "True":
            read_book = read_book + 1
    total_amount = len(display_statistic)
    if total_amount == 0:
        print("There is no book in your library")
    elif total_amount == 1:
        if read_book == 0:
            print("There is only one book in your library and it hasn't been read")
        else:
            print("There is only one book in your library and it has been read")
    else:
        percentage = read_book / total_amount *100
        print(f"There are {total_amount} books in your library and {percentage}% of them have been read")

# users choose a method
prompt = "WHat do you want to do? Enter the corresponding number\n" \
"Add a Book: 1\n" \
"Remove a Book: 2\n" \
"Search for a Book: 3\n" \
"Display All Books: 4\n" \
"Display Statistics: 5"
method = input(prompt).strip()
while not method in ["1", "2", "3", "4", "5"]:
    print("please enter the valid number")
    method = input(prompt).strip()

# add a book
if method == "1":
    def information():
        title = input("Please enter the book's title").strip().title()
        author = input("Please enter the book's author").strip().title()
        publication_year = input("Please enter the publication year")
        # check whether publication year is valid
        while not publication_year.isdigit():
            publication_year = input("publication year should be an integer, enter again")
        publication_year = int(publication_year)
        genre = input("Please enter the book's genre").strip().lower()
        read = input("Have you read the book? Enter read or unread").strip().lower()
        # turn read into boolean
        read_checker = False
        while read_checker == False:
            if read == "read":
                read_status = True
                read_checker = True
            elif read == "unread":
                read_status = False
                read_checker = True
            else:
                read = input("you should only enter read or unread").strip().lower()
        return title, author, publication_year, genre, read_status
    title, author, publication_year, genre, read_status = information()
    new_book = Library(title, author, publication_year, genre, read_status)    
    new_book.add_a_book()    


# remove a book
if method == "2":
    title_to_remove = input("Enter the book's title").strip().title()
    book_to_remove = Library(title_to_remove, None, None, None, None)
    book_to_remove.remove_a_book()

# search for a book
if method == "3":
    search_method = input("Search by title: 1 ,search by author: 2. Enter the corresponding number")
    if search_method == "1":
        search_title = input("Enter the title").strip().title()
        search_title_book = Library(search_title, None, None, None, None)
        search_title_book.search_for_a_book_title()
    elif search_method == "2":
        search_author = input("Enter the author").strip().title()
        search_author_book = Library(None, search_author, None, None, None)
        search_author_book.search_for_a_book_author()        

# display all books
if method == "4":
    display_all_books()

# display statistic
if method == "5":
    display_statistic()
```
