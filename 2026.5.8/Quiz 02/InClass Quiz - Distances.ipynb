### Distances

---

Read in the file __"distances.txt"__. It contains numbers with units meters (m), centimeters (cm) and millimeters (mm). Calculate the total distance in meters and print your result. i.e. convert all numbers into the same unit, cast them to floats and calculate the sum of all numbers.

```python
file = open (file = "distances.txt", mode = "r" )
distances = file.readlines()
data = []
for v in distances:
    x= v.split(" ")
    if x[1] == "cm\n" or x[1] == "cm":
        data.append(float(x[0])/100)
    elif x[1] == "mm\n" or x[1] == "mm":
        data.append(float(x[0])/1000)
    else:
        data.append(float(x[0]))
s = sum(data)
print(f"the sum of all numbers is {s}m")


```
